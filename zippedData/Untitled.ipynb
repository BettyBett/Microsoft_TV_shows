# Data Cleaning.

```python
# rename original_title to movie.
df1.rename(columns={'original_title': 'movie'}, inplace = True)
df1
```

```python
# merge the two tables df1 and df2 on the common column name 'movie'
df = pd.merge(df1,df2, on = 'movie',how = 'inner')
df
```

```python
# drop columns with insignificant information.
df = df.drop(columns = ['Unnamed: 0','release_date_y','id_x','title','genre_ids'])
df
```

```python
# Total number of null values in all the columns
df.isnull().sum()
```

```python
# Check for missing values in the table
df.isna().sum()
```

```python
# check for duplicate values
df.duplicated().sort_values().sum()
```

```python
# drop the duplicate values and keep the first
df.drop_duplicates(subset = ['movie'], keep = 'first', inplace = True)
```

```python
# counter-check on the duplicate values
df.duplicated().sort_values().sum()
```

```python
# view the table with no duplicates
df
```

```python
# remove the $ 'dollar sign' and ',' from the production_budget column
# convert production_budget to an integer
df['production_budget'] = df['production_budget'].replace({'\$':'',',':''},regex = True)
df['production_budget'].astype(str).astype(int)
```

```python
# remove the $ 'dollar sign' and ',' from the domestic_gross column
# convert domestic_gross to an integer
df['domestic_gross'] = df['domestic_gross'].replace({'\$':'',',':''},regex = True)
df['domestic_gross'].astype(str).astype(int)
```

1. Categorical Data

```python
#arrange the domestic gross in order from the highest to the least.
top_domestic_gross = df.sort_values(['domestic_gross'], ascending =[False])
top_domestic_gross.head(10)
```

```python
# arrange the worldwide_gross in order from the highest to the least 
top_worldwide_gross = df.sort_values(['worldwide_gross'], ascending =[False])
top_worldwide_gross.head(10)
```

```python
# extract the two columns production budget and domestic gross from the dataset
subs =df.iloc[:,7:9]
subs
```

```python
# convert the values in production budget and domestic gross to integers
subs['production_budget'] = subs['production_budget'].astype(int)
subs['domestic_gross'] = subs['domestic_gross'].astype(int)
```

```python
# Subtract the production budget from domestic gross to get the profit margin. 
df['profit_margin'] = subs['domestic_gross'] - subs['production_budget']
df['profit_margin']
```

```python
# arrange the profit margin from the highest 
df = df.sort_values(by ='profit_margin', ascending = False)
df.head(10)
```

```python
# description of the dataset
df.describe()
```

```python
# extract rows whereby the vote average, vote count and profit margin values are above the 75% percentile
df1 = df.loc[(df['vote_average'] > 6.9) & (df['vote_count'] > 2036) & (df['profit_margin']>17937050)]
df1.head(10)
```

# Data Visualization

```python
# plot a bar graph that shows the profit margins against the movie titles.
fig = plt.figure(figsize= (13,8))
movie = df1['movie'].head(10)
profit_margin = df1['profit_margin'].head(10)
# creating the bar graph
plt.bar(movie, profit_margin, width= 0.5)
plt.xlabel('movies')
plt.ylabel('profit_margin*(10^7)')
plt.title('The profit margins in the top 10 films')
plt.xticks(rotation = 90)
plt.show()
```

```python
# plot a bar graph that shows the production budget budget against the top 10 films
fig = plt.figure(figsize= (13,7))
movie = df1['movie'].head(10)
production_budget = df1['production_budget'].head(10)
# creating the bar graph
plt.bar(movie,production_budget, width= 0.6)
plt.xlabel('movie')
plt.ylabel('production_budget')
plt.title('Production budget against the top ten movies')
plt.xticks(rotation = 90)
plt.show()
```

```python
# graph on the relationship between the the vote average and worldwide gross 
# this determines how they relate
fig = plt.figure(figsize= (13,7))
ratings = df['vote_average'].head(20)
worldwide_gross = df['worldwide_gross'].head(20)
# creating the bar graph
plt.bar(worldwide_gross,ratings, width= 0.5)
plt.xlabel('worldwide_gross')
plt.ylabel('vote average')
plt.title('Ratings against worldwide gross')
plt.xticks(rotation = 90)
plt.show();
```
